# Feature Selection

We can try to find out how the features are correlated with the Target
and then remove other irrelevant features from the data. We use Cramer'V to calculate the correlation coefficient

```python
def cramers_V(var1,var2) :
  crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
  stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
  obs = np.sum(crosstab) # Number of observations
  mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
  return np.sqrt(stat/(obs*mini))

rows= []

for var1 in df_train:
  col = []
  for var2 in df_train:
    cramers = cramers_V(df_train[var1], df_train[var2]) # Cramer's V test
    col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V
  rows.append(col)

cramers_results = np.array(rows)
df_cramers = pd.DataFrame(cramers_results, columns = df_train.columns, index = df_train.columns)

df_cramers
```

```python
# Correlation heatmap using Cramer's V
plt.figure(figsize=(30,30))
mask = np.triu(np.ones_like(df_cramers.corr(), dtype=bool))
plt.title('Correlation Heatmap of DropOut Data')
a = sns.heatmap(df_cramers, square=True, annot=True, mask=mask, fmt='.3f', linecolor='black', cmap="Blues")
plt.show()
```

Based on the heatmap and the correlation coefficient, we can choose to eliminate variables that has coefficient with absolute value less that 0.1
